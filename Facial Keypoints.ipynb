```python
def impute(giv_col):
    mean_col = giv_col.mean()
    
    giv_col = giv_col.fillna(mean_col)
    return giv_col
```

```python
training_labels = training_labels.apply(lambda x: impute(x))
```

```python
#Lets Build the MOdel
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import Model
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data , val_data , train_label , val_label = train_test_split(training_images , train_labels , test_size=0.2 , random_state=4)
```

```python
train_data.shape
```

```python
val_data.shape
```

```python
#Model
input_vect = layers.Input(shape = [96,96,1])
x = layers.Conv2D(16 , 3 , activation='relu' , padding='same')(input_vect)
x = layers.MaxPool2D(2)(x)

x = layers.Conv2D(32 , 3 , activation='relu' , padding='same')(x)
x = layers.Conv2D(32 , 3 ,activation='relu' , padding='same' )(x)
x = layers.MaxPool2D(2)(x)


x = layers.Conv2D(64 , 3 , activation='relu' , padding='same')(x)
x = layers.MaxPool2D(2)(x)

x = layers.Conv2D(128, 3 , activation='relu' , padding='same')(x)
x = layers.MaxPool2D(2)(x)

x = layers.Conv2D(256, 3 , activation='relu' , padding='same')(x)
x = layers.MaxPool2D(2)(x)

x = layers.Conv2D(512, 3 , activation='relu' , padding='same')(x)
x = layers.MaxPool2D(2)(x)

x = layers.Flatten()(x)
x = layers.Dense(128 , activation='relu')(x)
x = layers.Dropout(0.2)(x)
output = layers.Dense(30, activation='relu')(x)
```

```python
model = Model(input_vect , output)
```

```python
model.summary()
```

```python
train_data = train_data / 255.0
val_data = val_data / 255.0
```

```python
train_label = train_label / 96.0
```

```python
val_label = val_label / 96.0
```

```python
model.compile(loss='mean_squared_error' , optimizer='adam' , metrics=['mse'])
```

```python
train_data = train_data.reshape(train_data.shape[0] , 96, 96 , 1)
val_data = val_data.reshape(val_data.shape[0] , 96 , 96 , 1)
model.fit(train_data , train_label , epochs=30 , verbose=1 , validation_data=(val_data , val_label))
```
